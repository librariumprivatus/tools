```python
import pathlib

import hashlib

import os



def cover_name(path):
    path = pathlib.Path(path)
    #print('path: ', path)

    
    # print('P ', path.as_posix())
    md5 = str(hashlib.md5(path.as_posix().encode()).hexdigest())
    #print(md5)
    
    from slugify import slugify

    
    path = path.with_name(slugify(path.stem) + path.suffix + '-' + str(md5[:8]))
    
    path = path.with_name(slugify(path.stem) + path.suffix + '-' + 'thumb' + '-' + str(md5[:8]))
    
    return path.name
    
# name1 = cover_name(path)
# print('NAME: ', name1)
```
