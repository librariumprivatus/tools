```python
import subprocess
import pathlib

def find_similar(file):
    #print(file.name)
    parent = file.parent
    for f in parent.iterdir():
        if f.suffix != '.pdf':
            continue
        
        if file.stem in f.name:
            # print('\t', 'Exist')
            # print('\t', 'pdf: ', f.name)
            return f
    return


def gen_djvu(file):
    
    method = '.ddjvu'

    
    pdf = file.with_stem(file.stem+method).with_suffix('.pdf')
    
    inp = file.as_posix()
    out = pdf.as_posix()
    
    
    # print('\t', 'inp: ', inp)
    # print('\t', 'out: ', out)
    
    ddjvue_com = ['ddjvu', '-format=pdf', f'{inp}', f'{out}']

    process = subprocess.Popen(ddjvue_com)
    process.wait()
    
    return pdf


def gen_cmd(file):
    
    method = '.ddjvu'

    
    pdf = file.with_stem(file.stem+method).with_suffix('.pdf')
    
    inp = file.as_posix()
    out = pdf.as_posix()
    
    
    # print('\t', 'inp: ', inp)
    # print('\t', 'out: ', out)
    
    ddjvue_com = ['ddjvu', '-format=pdf', f'{inp}', f'{out}']

    return ddjvue_com
    

def get_djvu_targets(files):
    targets = []
    for file in files:
        if file.suffix not in ['.djvu', '.djv']:
            continue
        
        similar = find_similar(file)
        if similar:
            continue
        
        targets.append(file)

    return targets


from concurrent.futures import ThreadPoolExecutor
import time

def exec_(cmd):
    '''
    single command run
    command must be prepared as subprocess.call
    '''
    ret = subprocess.call(cmd)
    if ret== 0:
        print("success...")
    else:
        print("error")


def pull_run(parallel_jobs, cmds):
    '''
    run pull of jobs
    input:
        parallel_jobs: integer, how many jobs at once
        cmds: list of commands
    '''
    with ThreadPoolExecutor(max_workers=parallel_jobs) as executor:
        futures = executor.map(exec_, cmds)

def gen_pdfs_from_djvu(files):
    for idx, file in enumerate(files):
        print(f'üê∏ {idx}: File: ', file.name)
        generated = gen_djvu(file)
        print('\t', generated)
        print('\n\n')

        
def gen_comds(files):
    cmds = []
    for idx, file in enumerate(files):
        cmd = gen_cmd(file)
        cmds.append(cmd)
    return cmds


def log_task(name, data=''):
    root = pathlib.Path(path_bibl)
    root = root.joinpath(name)

    f = open(root.as_posix(), "w")
    f.write("Success!")
    f.write("\n")
    for row in data:
        f.write(str(row))
    f.close()





```

```python

cmds = gen_comds(targets)
log_task(name='x-start.txt', data=cmds)

pull_run(100, cmds)

log_task(name='x-end.txt')
```
