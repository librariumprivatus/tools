```python
scripts = soup.find_all('script')

scr_text = ''
for script in scripts:
    data = script.text
    
    if 'ProjectData' in data:
        scr_text = data


print(scr_text)
```

```python
import json


def find_json_objects(text: str, decoder=json.JSONDecoder()):
    """Find JSON objects in text, and generate decoded JSON data"""
    pos = 0
    while True:
        match = text.find("{", pos)
        if match == -1:
            break
        try:
            result, index = decoder.raw_decode(text[match:])
            yield result
            pos = match + index
        except ValueError:
            pos = match + 1

found = list(find_json_objects(scr_text))
print(found)
# [{'json': 'objects'}, {'product': {'id': 1, 'name': 'first product'}}]
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By

def get_selenium_driver(url):
    options = webdriver.ChromeOptions()
    options.add_argument("headless")
    driver = webdriver.Chrome(options=options)
    driver.get(url)
    #print(driver.find_element(By.ID, 'wrapper').text)
    return driver


marc21_html_s = [] 
for id in data_ids[:2]:
    print(id)
    
    url = f'https://search.rsl.ru/ru/record/{id}'
    driver = get_selenium_driver(url)
    marc21_html = driver.find_element(By.ID, 'marc-rec').get_attribute('innerHTML')
    print(marc21_html)
    marc21_html_s.append(marc21_html)

    print()
```

```python
from bs4 import BeautifulSoup


def parse_rsl_macr21(html):
    soup = BeautifulSoup(html, 'html.parser')
    table = soup.find('table')
    trs = table.find_all('tr')
    if not trs:
        return None
    
    raw = {}
    for idx, tr in enumerate(table.find_all('tr')):
        # print(idx)
        
        
        tds = tr.find_all('td')
        if not tds:
            continue
        raw[str(tds[0].text)] = tds[1].text
    
    return raw
    

raw_marc21s = []   

for idx, marc21_html in enumerate(marc21_html_s):

    print('IDX: ', idx)
    data = parse_rsl_macr21(marc21_html)
    raw_marc21s.append(data)
    print(data)
    print()
    
    
```
